# Challenge 2 游뱤
Crear una tarjeta personal. El usuario (mediante input) puede elegir:

- Crear datos nuevos
- Modificar datos
- Eliminar datos

```python
tarjeta_personal = {}

while True:
    print("Elija una opcion numerica:")
    print("Opcion 1: Crar datos nuevos")
    print("Opcion 2: Modificar dato")
    print("Opcion 3: Eliminar dato")
    print("Opcion 4: Salir \n")
    opcion = (input("Ingrese el numero de la opcion:"))

    if opcion == "1":
        print("Agregar datos a tarjeta personal")
        llave = input("Ingrese la Clave:")
        valor = input("Ingrese el valor")
        tarjeta_personal[llave] = valor
        print(f"creada la clave {llave} con el valor {valor}\n")

    elif opcion == "2":
        print("Modificar datos de tarjeta personal")
        llave = input("Ingrese la Clave a modificar:")
        valor = input("Ingrese el nuevo valor")

        if llave in tarjeta_personal:
            tarjeta_personal[llave] = valor
            print(f"Modificada la clave {llave} con el valor {valor}\n")
        else: 
            print(f"clave {llave} no existente\n")

    elif opcion == "3":
        print("Eliminar datos de tarjeta personal")
        llave = input("Ingrese la Clave a eliminar:")

        if llave in tarjeta_personal:
            del tarjeta_personal[llave]
            print(f"Eliminaste la clave {llave} de la tarjeta personal\n")
        else:
            print(f"clave {llave} no existente\n")

    elif opcion == "4":
        print("Saliendo del programa, tarjeta personal finalizada\n")
        print(tarjeta_personal)
        break

    else:
        print("dato invalido, recuerde solo seleccionar 1, 2, 3 o 4\n")

```

# Challenge 3 游뱤
Crear una agenda de contactos. El usuario puede elegir:

1. Mostrar los contactos
2. Crear contacto
3. Modificar datos del contacto

```python
agenda_contactos = {}

while True:
    datos = {}
    print("Elija una opcion numerica:")
    print("Opcion 1: Mostrar contactos")
    print("Opcion 2: Crear contactos")
    print("Opcion 3: Modificar datos de contacto")
    print("Opcion 4: Salir \n")
    opcion = (input("Ingrese el numero de la opcion:"))

    if opcion == "1":
        print("Datos de contactos")
        print(f" {agenda_contactos} \n")

    elif opcion == "2":
        print("Creando contacto")
        nombre = input("Ingrese nombre del contacto")
        llave = input("Ingrese una clave del contacto")
        valor = input("Ingrese un valor a la clave")
        datos[llave] =  valor
        agenda_contactos[nombre] = datos
        print(f"creado el contacto {nombre} con la clave {llave} y el valor {valor}\n")


    elif opcion == "3":
        print("Modificando datos de contacto")
        print(agenda_contactos.keys())
        nombre = input("Ingrese nombre del contacto que desea modificar")
        llave = input("Ingrese una clave del contacto que quiere modificar")
        valor = input("Ingrese un valor a la clave")
        if nombre in agenda_contactos and llave in agenda_contactos.values():
            datos[llave] =  valor
            agenda_contactos[nombre] = datos
            print(f"Modificado el contacto {nombre} en la clave {llave} con el valor {valor}\n")
        else:
            print("Datos invalidos")

    elif opcion == "4":
        print("Saliendo del programa, agenda finalizada\n")
        print(agenda_contactos)
        break
    
    else: 
        print("dato invalido, recuerde solo seleccionar 1, 2, 3 o 4\n")
```

# Challenge 4游뱤
Crear dos m칠todos para nuestro animal y ejecutar, debe ser algo espec칤fico de ese animal.

    Ejemplo: "Cuando hablo, digo X".
    Ejemplo: "Cuando me muevo, hago X".

```python
class Perro:

    def __init__(self, due침o, ladrido, movimiento):
        self.due침o = due침o
        self.ladrido = ladrido
        self.movimiento = movimiento

    def ladrar(self):
        return self.ladrido
    
    def seguir(self):
        return f"{self.movimiento} con mi due침o {self.due침o}"
    
perri = Perro("Ale", "guau guau", "haciendo la croqueta")

print(perri.ladrar())
print(perri.seguir())
```

# Challenge 5 游뱤
Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.

    Observaci칩n: Editar las clases hijas para que no se inicialicen.

```python
class Animal:

    def __init__(self, color, raza):
        self.color = color
        self.raza = raza

class Perro(Animal):

    def ladrar(self):
        return "guau guau"
    
class Gato(Animal):

    def maullar(self):
        return "miau miau"
    
perri = Perro("Negro", "York")
gati = Gato("Naranja", "Calle")

print(f"El perro es de color {perri.color}, de la raza {perri.raza} y hace {perri.ladrar()}")
print(f"El gato es de color {gati.color}, de la raza {gati.raza} y hace {gati.maullar()}")
```
